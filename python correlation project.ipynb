```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts configuration of plots created

# read in data
df = pd.read_csv(r'C:/Users/tjean/OneDrive/Documents/movies.csv.zip')
```

```python
#success
#look at data
df.head()
```

```python
# see if there is missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings in Hundred Millions')
plt.ylabel('Budget for Film in Hundred Millions')
plt.show()
```

```python
# regression plot of budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
# shows positive correlation between budget and gross
```

```python
#determining the actual correlation
#pearson method below (default) - others are kendall and spearman
df.corr()
```

```python
correlation_matrix = df.corr(method='spearman')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Between Numeric Values')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='spearman')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Between Numeric Values')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
